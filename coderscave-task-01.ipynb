```python
new_working_data = pd.read_csv('Hackathon_Working_Data.csv') #read
```

```python
#REMOVE NULL VALUES
```

```python
# Filtering the DataFrame to exclude rows where 'PRICE', 'VALUE', and 'BILL_AMT' columns are equal to zero
new_working_data = new_working_data[
    (new_working_data['PRICE'] != 0) & 
    (new_working_data['VALUE'] != 0) & 
    (new_working_data['BILL_AMT'] != 0)
]

# Printing the filtered DataFrame
print(new_working_data)
```

```python
#REMOVE DUPLICATE VALUES
new_working_data = new_working_data.loc[~new_working_data.duplicated()]
print(new_working_data)
```

```python
columns_to_remove = ['SGRP', 'SSGRP']  # Adding coloumns

# Drop the specified columns
data_without_columns = new_working_data.drop(columns=columns_to_remove)
print(data_without_columns)
```

```python
# Counting the occurrences of each day in the 'DAY' column and creating a bar plot
ax = data_without_columns['DAY'].value_counts().plot(kind="bar", title='Highest Billed day')

# Setting the label for the x-axis
ax.set_xlabel("DAY")

# Setting the label for the y-axis
ax.set_ylabel("COUNT")
```

```python
# Importing the matplotlib library
import matplotlib.pyplot as plt

# Counting the occurrences of each month in the 'MONTH' column
month_counts = data_without_columns['MONTH'].value_counts()

# Creating a new figure and setting its size
plt.figure(figsize=(7, 5))

# Creating a pie chart with month_counts as data, month labels as labels, 
# displaying percentages with one decimal point, and starting the angle at 140 degrees
plt.pie(month_counts, labels=month_counts.index, autopct='%1.1f%%', startangle=140)

# Setting the title of the pie chart
plt.title('Distribution of Months')

# Displaying the pie chart
plt.show()
```

```python
#TOP SOLD STORE
dx = new_working_data['STORECODE'].value_counts()\
.plot(kind='bar',title = 'Top selled store',color='Orange')
dx.set_xlabel('Store name')
dx.set_ylabel('Count')
```

```python
#PLOTTING FOR DAY AND THE BILL AMOUNT
```

```python
xaxis=new_working_data['DAY'].astype(int)
yaxis=new_working_data['BILL_AMT'].astype(int)

barlist=plt.bar(xaxis,yaxis,color='blue')

# Adding labels to the x-axis and y-axis
plt.xlabel('Day')
plt.ylabel('Total Bill Amount')

# Displaying the plot
plt.show()
```

```python
# Load your DataFrame from a file if needed
# If df is already loaded or defined elsewhere, this line assigns it to new_working_data
df = new_working_data

# Extract numerical part from 'STORECODE' and convert it to int
df['STORECODE'] = df['STORECODE'].astype(str).str.extract('(\d+)').astype(int)

# Aggregate bill amounts by store code
store_sales = df.groupby('STORECODE')['BILL_AMT'].sum().reset_index()

# Creating a new figure with a specified size
plt.figure(figsize=(6, 6))

# Creating a bar plot to visualize total bill amount for each store code
plt.bar(store_sales['STORECODE'], store_sales['BILL_AMT'], color='maroon')

# Labeling the x-axis
plt.xlabel('Store Code')

# Labeling the y-axis
plt.ylabel('Total Bill Amount')

# Adding a title to the plot
plt.title('Total Bill Amount by Store Code')

# Displaying the plot
plt.show()
```

```python
# Calculate total sales by group (brand)
brand_sales = df.groupby('GRP')['BILL_AMT'].sum().reset_index()

# Display the total bill amount by brand
print(brand_sales)

# Sort the DataFrame by total bill amount in descending order and select the top 15 groups
top_groups = brand_sales.sort_values(by='BILL_AMT', ascending=False).head(15)

# Plot the bar chart for the top 15 groups
plt.figure(figsize=(10, 6))
plt.bar(top_groups['GRP'], top_groups['BILL_AMT'], color='green')
plt.xlabel('Group')
plt.ylabel('Total Bill Amount')
plt.title('Total Bill Amount by Group (Top 15)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Calculate total quantity and total price by brand
brand_totals = df.groupby('GRP').agg({'QTY': 'sum', 'PRICE': 'sum'}).reset_index()

# Sort the DataFrame by total quantity or total price in descending order and select the top 30 brands
top_companies_qty = brand_totals.sort_values(by='QTY', ascending=False).head(30)
top_companies_price = brand_totals.sort_values(by='PRICE', ascending=False).head(30)

# Plot total quantity and total price by brand for the top 30 brands
plt.figure(figsize=(12, 6))

# Plot total quantity
plt.bar(top_companies_qty['GRP'], top_companies_qty['QTY'], color='red', alpha=0.7, label='Total Quantity')

# Plot total price
plt.bar(top_companies_price['GRP'], top_companies_price['PRICE'], color='pink', alpha=0.7, label='Total Price')

plt.xlabel('Brand')
plt.ylabel('Total Quantity/Price')
plt.title('Total Quantity and Price by Top 30 Brands')
plt.xticks(rotation=90)  
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# CORRELATION GRAPH FOR QUANTITY AND PRICE
# - POSITIVE correlation: As the quantity of products sold increases, the price tends to increase.
# - NEGATIVE correlation: As the quantity of products sold increases, the price tends to decrease.
```

```python
brand_correlation = df.groupby('GRP')[['QTY', 'PRICE']].corr().iloc[0::2,-1].reset_index()

# Plot the correlation coefficients
plt.figure(figsize=(14, 6))
sns.barplot(x='GRP', y='PRICE', data=brand_correlation, palette='coolwarm')
plt.xlabel('Brand')
plt.ylabel('Correlation Coefficient')
plt.title('Correlation between Quantity and Price of Brands')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
